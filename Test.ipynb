### This Program is give you longest commom string in both strings.

```python
x = "abbacdcba"
y = "bcdbbcaa"
s = ""
l = [[0 for x in range(len(y)+1)] for x in range(len(x)+1)]

for i in range(len(x)+1):
        for j in range(len(y)+1):
            if i == 0 or j == 0:
                l[i][j] = 0
            elif x[i-1] == y[j-1]:
                l[i][j] = l[i-1][j-1] + 1
            else:
                l[i][j] = max(l[i-1][j], l[i][j-1])
anslen = l[len(x)][len(y)]
ans = [""] * (anslen + 1)
ans[anslen] = ""
i = len(x)
j = len(y)
while i > 0 and j > 0:
    if x[i-1] == y[j-1]:
            ans[anslen - 1] = x[i - 1]
            i -= 1
            j -= 1
            anslen -= 1

    elif l[i-1][j] > l[i][j-1]:
        i -= 1
    else:
        j -= 1
for i in ans:
    if i != "":
        s = s + i
print(s)
```
